Información de relación cargo laboral vs años de experiencia y expectativa salarial

```python
data = {
    'Puesto': ['Desarrollador Junior', 'Desarrollador Junior', 'Desarrollador Senior', 'Ingeniero de Datos', 'Científico de Datos', 'Desarrollador Backend',
                          'Desarrollador Frontend', 'Ingeniero DevOps', 'Desarrollador Móvil', 'Analista de QA'],
    'Ciudad': ['Bogotá', 'Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Cali', 'Medellín', 'Pereira', 'Pereira', 'Medellín'],
    'Experiencia': ['1-3 años', '3-5 años', 'Mayor a 5 años', '3-5 años', '3-5 años', 'Mayor a 5 años', '1-3 años', 'Mayor a 5 años', '1-3 años', '3-5 años'],
    'Salario Anual (COP)': [30000000, 45000000, 80000000, 65000000, 70000000, 75000000, 40000000, 85000000, 35000000, 50000000]
}

df = pd.DataFrame(data)

print(df)
df
```

Generación de valor de salario mensual

Cáluculo del salario promedio según ciudad (creo se prodía mejorar con analisis de años de experiencia en los distintos campos)

```python
def SalarioPromedio(df, ciudad):
    filtro = (df['Ciudad'] == ciudad)
    df_filtrado = df[filtro]

    # Calcular el salario promedio
    if not df_filtrado.empty:
        salario_promedio = df_filtrado['Salario Anual (COP)'].mean()
        return salario_promedio
    else:
        return None
```

Ejecución de la funcion

```python
ciudad = 'Cali'
SalarioPromedio(df, ciudad)
```

Ejecución de la funcion para todas las ciudades que tenga el arreglo

```python
ciudades = df['Ciudad'].unique()
for ciudad in ciudades:
  salario_promedio= SalarioPromedio(df, ciudad)
  print('En la ciudad de ' + ciudad+ ', el salario promedio es: ' + str(salario_promedio))
```
