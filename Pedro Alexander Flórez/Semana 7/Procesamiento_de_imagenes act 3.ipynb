Rotacion de la imagen.

```python
image = cv2.imread('imagen1.jpg')

height, width = image.shape[:2]

center = (width/2, height/2)

angle = 45
M = cv2.getRotationMatrix2D(center, angle, 1.0)

rotated = cv2.warpAffine(image, M, (width, height))

cv2_imshow(rotated)
```

Traslacion de imagenes

```python
tx, ty = 100, 50
M = np.float32([[1, 0, tx], [0, 1, ty]])

#aplicar la translacion

#translated = cv2.warpAffine(image, M, image.shape[1], image.shape[0])

#Error al representar el tama√±o de la imagen suguerencia realizada por gemini

translated = cv2.warpAffine(image, M, (width, height))

cv2_imshow(translated)
```

Escala de imagenes

```python
new_width, new_height = 400, 200

#aplicar la nueva escala de la imagen

scaled = cv2.resize(image, (new_width, new_height))

cv2_imshow(scaled)
```

Suavizado de imagenes

```python
smoothed = cv2.GaussianBlur(image, (5,5),0)

cv2_imshow(smoothed)
```

Realce de imagenes

```python
kernel = np.array([[-1, -1, -1],
                   [-1, 9, -1],
                   [-1, -1, -1]])
sharpened = cv2.filter2D(image, -1, kernel)

cv2_imshow(sharpened)
```

Detenci√≥n de bordes de imagenes

```python
image = cv2.imread('imagen1.jpg', cv2.IMREAD_GRAYSCALE)

sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 5)

edges = cv2.magnitude(sobelx, sobely)

edges = cv2.normalize(edges, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

cv2_imshow(edges)
```
