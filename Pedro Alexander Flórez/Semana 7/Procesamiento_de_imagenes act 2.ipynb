```python
camera = skimage.data.camera()

print(type(camera))
print(camera.shape)

#Mostrar la imagen
skimage.io.imshow(camera)
skimage.io.show()
```

Carga de imagen descarg√°da

```python
image_path = 'imagen1.jpg'

image = skimage.io.imread(image_path)

skimage.io.imshow(image)
skimage.io.show()
```

Rotacion de imagenes

```python
rotated_image = skimage.transform.rotate(image, angle = -30)

skimage.io.imshow(rotated_image)
skimage.io.show()
```

Suavizaci√≥n de imagen

```python
smoothed_image = skimage.filters.gaussian(image, sigma=2)

skimage.io.imshow(smoothed_image)
skimage.io.show()
```

Filtro Sobel

```python
sobel_image = skimage.filters.sobel(image)

skimage.io.imshow(sobel_image)
skimage.io.show()
```
