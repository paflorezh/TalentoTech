Importacion de las librerias

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

Realización de convolución con OpenCV

```python
image = cv2.imread('imagen1.jpg')

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

convolved_image = cv2.filter2D(image, -1, kernel)

print("Imagen Original: ")
cv2_imshow(image)
print("-----------------------------")
print("Imagen con la convolución: ")
cv2_imshow(convolved_image)
```

Cambios en el kernel para ver comportamiento

```python
kernel = np.array([[0, -1, 0],
                   [-1, 4.5, -1],
                   [0, -1, 0]])

convolved_image = cv2.filter2D(image, -1, kernel)

print("Imagen Original: ")
cv2_imshow(image)
print("-----------------------------")
print("Imagen con la convolución: ")
cv2_imshow(convolved_image)
```
