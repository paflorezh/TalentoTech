Importacion de las librerias

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#Carga del modelo

faceClassif = cv2.CascadeClassifier('gatos.xml')
image = cv2.imread('gaticos.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = faceClassif.detectMultiScale(gray, scaleFactor=1.1,
                                     minNeighbors=5,
                                     minSize=(30,30),
                                     maxSize=(200,200))
print(faces)
```

```python
#Dibujo de rectangulos en los rostros

for(x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w, y+h), (0,255,0),2)

cv2_imshow(image)
```
